```python
from gluonts.dataset import common
from gluonts.mx.model import deepar
from gluonts.mx.trainer import Trainer
import pandas as pd
import matplotlib.pyplot as plt
from gluonts.dataset.common import ListDataset
# 读入数据
df = pd.read_table("LD2011_2014.txt", sep=";", index_col = 0, parse_dates=True,decimal=',')
# 数据集前10条时间序列可视化
fig, axs = plt.subplots(5, 2, figsize = (20, 10))
for i in range(0, 10):
    row = int(i/2)
    col = i % 2
    df[df.columns[i]].loc["2014-12-01":"2014-12-14"].plot(ax=axs[row][col])
    axs[row][col].set_xlabel("date")
    axs[row][col].set_ylabel("kW Consumption")
    axs[row][col].grid(which = "minor", axis='x')
```

```python
#---------------------
# 建立训练和测试数据集
#---------------------
# 对于模型测试来说，这里整个数据集过大，训练所需资源较大，为节省训练时间，我们只选取前面8条曲线来训练
# 1. 首先需要对每个曲线添加cat，这里我们假设每条曲线都是相互独立的
df = df.reset_index(drop=True).T.reset_index()
ts_code = df["index"].astype("category").cat.codes.values # 每条曲线的cat
# 2. 分割训练集和测试集
num_train_samples = 135000
df_train = df.iloc[:, 1:num_train_samples].values # 从每条曲线选取前n_train_samples个样本
df_test = df.iloc[:, num_train_samples+1:].values
# 3. 设置数据集的metadata
freq = "15min"
start_train = pd.Timestamp("2011-01-01 00:15:00", freq = freq)
start_test = pd.Timestamp("2014-11-07 05:30:00", freq = freq)
prediction_length = 672
# 4. 构筑训练集和测试集
num_series = 8
train_dataset = ListDataset([
    {
        trainTa.TARGET: target,
        trains.START: start_train,
        traine.FEAT_STATIC_CAT: fsc
    } for(target, fsc) in zip(df_train[0:num_series], ts_code[0:num_series].reshape(-1,1))
], freq = freq)

test_dataset = ListDataset([
    {
        FieldName.TARGET: target,
        FieldName.START: start_test,
        FieldName.FEAT_STATIC_CAT: fsc
    } for(target, fsc) in zip(df_test[0:num_series], ts_code[0:num_series].reshape(-1,1))
], freq = freq)
```

```python
# 建立模型
estimator = DeepAREstimator(freq = freq,
                           context_length = 672, # RNN窗口长度
                           prediction_length = prediction_length, # 预测序列长度
                           use_feat_static_cat = True, # 注意需要使用我们的类别数据
                           cardinality = [1], # 每个类别属性包含的类标签长度，当use_feat_static_cat为True时必须要设置这一个参数
                           num_layers = 2, # 双层RNN
                           num_cells = 32, # 每层包含32个LSTM cell
                           cell_type = "lstm",
                           trainer=Trainer(epochs=5) # 训练5代
                           )
# 训练模型
predictor = estimator.train(training_data=train_dataset)
```

```python
# 进行预测
forcast_it, ts_it = make_evaluation_predictions(
    dataset = test_dataset,
    predictor = predictor,
    num_samples = 100,
)
forecasts = list(forcast_it)
tss = list(ts_it)

# 对结果进行可视化
def plot_prob_forecasts(ts_entry, forecast_entry):
    plot_length = 1000
    prob_intervals = (50.0, 95.0)
    legend = ["True values", "median prediction"] + [f"{k}% prob interval" for k in prob_intervals][::-1]

    fig = plt.figure(figsize = (10, 5), dpi = 120)
    ax = plt.gca()
    ts_entry[-plot_length:].plot(ax = ax)
    forecast_entry.plot(prediction_intervals = prob_intervals, color = 'g')
    plt.legend(legend)
    plt.show()

for i in range(num_series):
    plot_prob_forecasts(tss[i], forecasts[i], i)
```
